```python
import pygame
import random
import time


pygame.init()

GAME_WIDTH = 720
GAME_HEIGHT = 480

SPEED = 20
OBJECT_SIZE = 10
SNAKE_LENGTH = 1
SCORE = 0

SNAKE_COLOR = 'GREEN'
FOOD_COLOR = 'RED'
BACKGROUND_COLOR = 'BLACK'



game_display = pygame.display.set_mode((GAME_WIDTH, GAME_HEIGHT))
pygame.display.set_caption('Snake Game')

fps = pygame.time.Clock()

font = pygame.font.SysFont('times new roman', 20)


def draw_snake(snake_size, snake_body):
    for p in snake_body:
        pygame.draw.rect(game_display, SNAKE_COLOR, [p[0], p[1], snake_size, snake_size])
        
        
def show_score():
    text = font.render('Score: ' + str(SCORE), True, 'cyan')
    game_display.blit(text, [0,0])
    
    
def game_over():
    game_display.fill(BACKGROUND_COLOR)
    close_window = font.render('Your Score is: ' + str(SCORE), True, 'white')
    game_display.blit(close_window, [GAME_HEIGHT/2, GAME_WIDTH/2])
    pygame.display.flip()

    time.sleep(2)
    
    pygame.quit()
    quit()
    
    
def start_screen():
    game_display.fill((210,210,210))
    play = font.render('Press any key to continue', True, 'BLACK')
    game_display.blit(play, [GAME_HEIGHT/2, GAME_WIDTH/2])
    pygame.display.flip()
    
    s = False
    while not s:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                quit()
                
            if event.type == pygame.KEYDOWN:
                s = True
    return s
            
    
    
    
def run_game():
    global SNAKE_LENGTH, SCORE
    
#     x = GAME_WIDTH / 2
    x = 0
    y = GAME_HEIGHT / 2
    
    x_move = 0
    y_move = 0
    
    direction = 'RIGHT'
    change_direction = 'RIGHT'
    
    snake_body = []
  
    food_x = random.randint(1, (GAME_WIDTH / OBJECT_SIZE)) * OBJECT_SIZE
    food_y = random.randint(1, (GAME_HEIGHT / OBJECT_SIZE)) * OBJECT_SIZE
    
    start = start_screen()
    
    while start:
        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                game_over()
                break
                
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT:
                    change_direction = 'LEFT'
                if event.key == pygame.K_RIGHT:
                    change_direction = 'RIGHT'
                if event.key == pygame.K_UP:
                    change_direction = 'UP'
                if event.key == pygame.K_DOWN:
                    change_direction = 'DOWN'

                    
        if change_direction == 'UP' and direction != 'DOWN':
            direction = 'UP'
            x_move = 0
            y_move = -OBJECT_SIZE
        if change_direction == 'DOWN' and direction != 'UP':
            direction = 'DOWN'
            x_move = 0
            y_move = OBJECT_SIZE
        if change_direction == 'RIGHT' and direction != 'LEFT':
            direction = 'RIGHT'
            x_move = OBJECT_SIZE
            y_move = 0
        if change_direction == 'LEFT' and direction != 'RIGHT':
            direction = 'LEFT'
            x_move = -OBJECT_SIZE
            y_move = 0
            
                        
        if (x >= GAME_WIDTH) or (x < 0) or (y >= GAME_HEIGHT) or (y < 0):
            game_over()
            break
            
        x += x_move
        y += y_move
        
        game_display.fill(BACKGROUND_COLOR)
        pygame.draw.circle(game_display, FOOD_COLOR, [food_x, food_y], OBJECT_SIZE//2)
        
        snake_body.append([x,y])
        
        if len(snake_body) > SNAKE_LENGTH:
            del snake_body[0]
            
        for body in snake_body[:-1]:
            if body == [x,y]:
                game_over()
        
          
        draw_snake(OBJECT_SIZE, snake_body)
        show_score()

        
        pygame.display.update()
        
        if (x == food_x) and (y == food_y):
            food_x = random.randint(1, (GAME_WIDTH / OBJECT_SIZE)) * OBJECT_SIZE
            food_y = random.randint(1, (GAME_HEIGHT / OBJECT_SIZE)) * OBJECT_SIZE
            SNAKE_LENGTH += 1
            SCORE += 1
            
        fps.tick(SPEED)

    
run_game()
```
